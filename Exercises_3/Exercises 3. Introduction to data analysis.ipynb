## Problem 1 - Read the data and remove NaN values

The first step for this problem is to read the data file 6153237444115dat.csv into a variable data using pandas and cleaning it a bit:
 - Select the columns USAF, YR--MODAHRMN, TEMP, MAX, MIN from the data DataFrame and assign them to a variable selected
 - Remove all rows from selected that have NoData in the column TEMP using the dropna() function

```python
workdata_cropped = workdata[["USAF", "YR--MODAHRMN", "TEMP", "MAX", "MIN"]].copy()
workdata_cropped = workdata_cropped.dropna(subset=["TEMP"])
print("Is nans in TEMP? - ", workdata_cropped["TEMP"].hasnans)
workdata_cropped.head()
```

### Problem 2 - Convert temperatures to Celsius

Convert the temperature values from Fahrenheits to Celsius:
 - Create a new column to selected called Celsius.
 - Convert the Fahrenheit temperatures from TEMP using the conversion formula below and store the results in the new Celsius column:
   - TempCelsius = (TempFahrenheit - 32) / 1.8
 - Round the values in the Celsius column to have 0 decimals (do not create a new column, update the current one)
 - Convert the Celsius values into integers (do not create a new column, update the current one)

```python
workdata_cropped["CELSIUS"] = 0.0
workdata_cropped["CELSIUS"] = (workdata_cropped["TEMP"] - 32) / 1.8
workdata_cropped["CELSIUS"] = workdata_cropped["CELSIUS"].round()
workdata_cropped["CELSIUS"] = workdata_cropped["CELSIUS"].astype(int)
print("CELSIUS type is", workdata_cropped["CELSIUS"].dtype)
workdata_cropped.head()


# Если формула была бы более сложной, то следовало бы воспользоваться apply()
```

### Problem 2 - Select data and compare temperatures between months

The median temperatures above consider data from the entire summer (May-Aug), hence the differences might not be so clear. Let’s now find out the mean temperatures from May and June 2017 in Kumpula and Rovaniemi:
 - From the kumpula and rovaniemi DataFrames, select the rows where values of the YR--MODAHRMN column are from May 2017. Assign these selected rows into the variables kumpula_may and rovaniemi_may
 - Repeat the procedure for the month of June and assign those values into variables to kumpula_june and rovaniemi_june
 - Calculate and print the mean, min and max Celsius temperatures for both places in May and June using the new subset dataframes (kumpula_may, rovaniemi_may, kumpula_june, and rovaniemi_june). Answer to following questions:
   - Does there seem to be a large difference in temperatures between the months?
   - Is Rovaniemi a much colder place than Kumpula?

```python
kumpula_data["YR--MODAHRMN"] = pandas.to_datetime(kumpula_data["YR--MODAHRMN"], format='%Y%m%d%H%M')
rovaniemi_data["YR--MODAHRMN"] = pandas.to_datetime(rovaniemi_data["YR--MODAHRMN"], format='%Y%m%d%H%M')
kumpula_may = kumpula_data[kumpula_data["YR--MODAHRMN"].dt.month == 5]
rovaniemi_may = rovaniemi_data[rovaniemi_data["YR--MODAHRMN"].dt.month == 5]
kumpula_june = kumpula_data[kumpula_data["YR--MODAHRMN"].dt.month == 6]
rovaniemi_june = rovaniemi_data[rovaniemi_data["YR--MODAHRMN"].dt.month == 6]
print("Mean Celsius temperatures in Kumpula in May 2017", kumpula_may["CELSIUS"].mean())
print("Min Celsius temperatures in Kumpula in May 2017",  kumpula_may["CELSIUS"].min())
print("Max Celsius temperatures in Kumpula in May 2017",  kumpula_may["CELSIUS"].max())
print()
print("Mean Celsius temperatures in Rovaniemi in May 2017", rovaniemi_may["CELSIUS"].mean())
print("Min Celsius temperatures in Rovaniemi in May 2017",  rovaniemi_may["CELSIUS"].min())
print("Max Celsius temperatures in Rovaniemi in May 2017",  rovaniemi_may["CELSIUS"].max())
print()
print("Mean Celsius temperatures in Kumpula in June 2017", kumpula_june["CELSIUS"].mean())
print("Min Celsius temperatures in Kumpula in June 2017",  kumpula_june["CELSIUS"].min())
print("Max Celsius temperatures in Kumpula in June 2017",  kumpula_june["CELSIUS"].max())
print()
print("Mean Celsius temperatures in Rovaniemi in June 2017", rovaniemi_june["CELSIUS"].mean())
print("Min Celsius temperatures in Rovaniemi in June 2017",  rovaniemi_june["CELSIUS"].min())
print("Max Celsius temperatures in Rovaniemi in June 2017",  rovaniemi_june["CELSIUS"].max())
```

In this problem you should:
 - Summarize the information for each day by aggregating (grouping) the DataFrame using the groupby() function.
